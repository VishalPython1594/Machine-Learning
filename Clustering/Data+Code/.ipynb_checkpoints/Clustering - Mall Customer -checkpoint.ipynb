```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
import warnings
warnings.filterwarnings('ignore')
import sklearn
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:,3:].values
```

```python
x
```

# K-Means Cluster

```python
from sklearn.cluster import KMeans
```

```python
# To find the best K (Cluster) basis Elbow method

wcss = []

for i in range(1,15):
    kmeans = KMeans(n_clusters=i,init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 15), wcss)
```

```python
wcss
```

```python
# As per elbow method, best k value is 5

kmeans = KMeans(n_clusters=5, init='k-means++')
y_kmeans = kmeans.fit_predict(x)
```

```python
kmeans.inertia_
```

```python
y_kmeans
```

```python
combind = pd.concat([dataset, pd.DataFrame(y_kmeans)], axis=1)
combind
```

```python
combind.tail(20)
```

```python
# x = Annual Income (k$)  and Spending Score (1-100)
```

```python
# visualize the cluster

plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s=100, c='red',label='Cluster_0')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s=100, c='blue',label='Cluster_1')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1], s=100, c='yellow',label='Cluster_2')
plt.scatter(x[y_kmeans==3,0], x[y_kmeans==3,1], s=100, c='pink',label='Cluster_3')
plt.scatter(x[y_kmeans==4,0], x[y_kmeans==4,1], s=100, c='green',label='Cluster_4')
```

```python
combind.to_csv("New_Cluster_data.csv")
```

# Hierarchical Clustering

```python
dataset.head()
```

```python
x = dataset.iloc[:,3:].values
x
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(x, method = 'ward'))
plt.title("Dendrogram for Clustering")
plt.xlabel("Customer")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering( n_clusters=5,affinity='euclidean',linkage='ward')
hc
```

```python
y_hc = hc.fit_predict(x)
y_hc
```

```python
cmbind = pd.concat([dataset, pd.DataFrame(y_hc)], axis=1)
```

```python
cmbind
```

```python
plt.scatter(x[y_hc==0,0], x[y_hc==0,1], c='black', label = 'H-Cluster_0')
plt.scatter(x[y_hc==1,0], x[y_hc==1,1], c='red', label = 'H-Cluster_1')
plt.scatter(x[y_hc==2,0], x[y_hc==2,1], c='green', label = 'H-Cluster_2')
plt.scatter(x[y_hc==3,0], x[y_hc==3,1], c='yellow', label = 'H-Cluster_3')
plt.scatter(x[y_hc==4,0], x[y_hc==4,1], c='pink', label = 'H-Cluster_4')
```
