```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('German_Credit_Data.csv')
df.head()
```

```python
x = df.drop('status',axis=1)
y = df['status']
```

```python
x = pd.get_dummies(x,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
mod = GradientBoostingClassifier()
```

```python
params = [{'n_estimators':[100,200,300,400,500], 'learning_rate':[0.01,0.1,0.5,0.7,0.9]}]
mod_cv = GridSearchCV(mod,params,cv=10,scoring='roc_auc')
mod_cv.fit(x_train,y_train)
```

```python
mod_cv.best_score_
```

```python
mod_cv.best_params_
```

### Performance Measures - Gradient Bossting

```python
pred_y = mod_cv.predict(x_test)
```

```python
from sklearn import metrics
```

```python
mat = metrics.confusion_matrix(y_test,pred_y)
mat
```

```python
print(metrics.classification_report(y_test,pred_y))
```

```python
auc = metrics.roc_auc_score(y_test, mod_cv.predict_proba(x_test)[:,1])
auc
```

```python
auc_1 = metrics.roc_auc_score(y_test,pred_y)
auc_1
```

```python
fpr,tpr,_ = metrics.roc_curve(y_test,mod_cv.predict_proba(x_test)[:,1])
auc = metrics.auc(fpr,tpr)
```
