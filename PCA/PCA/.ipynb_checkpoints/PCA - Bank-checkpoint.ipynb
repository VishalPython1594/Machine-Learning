```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('bank.csv')
```

```python
df.head()
```

```python
pd.set_option('display.max_columns',30)
```

```python
df.describe().T
```

```python
df.subscribed.value_counts()
```

```python
#IMBALANCED DATA
```

```python
from sklearn.utils import resample 
```

```python
df_subs_no = df[df['subscribed'] == 'no']
df_subs_yes = df[df['subscribed'] == 'yes']
```

```python
df_subs_no
```

```python
df_subs_yes
```

```python
df_minority_upsampled = resample(df_subs_yes,replace=True,n_samples=4000,random_state=42)
```

```python
new_df = pd.concat([df_subs_no,df_minority_upsampled])
```

### PCA with Whiten = True

```python
new_df.subscribed.value_counts()
```

```python
x = new_df.drop('subscribed',axis=1)
y = new_df['subscribed']
```

```python
x_dum = pd.get_dummies(x,drop_first=True,dtype='int')
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics as m
from sklearn.preprocessing import StandardScaler
```

```python
pca = PCA(n_components=0.95,whiten=True)
```

```python
mod_pc = pca.fit_transform(x_dum)
```

```python
pca.n_components_
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
new_df.subscribed = new_df.subscribed.apply(lambda x : 1 if x == 'yes' else 0)
```

```python
y = new_df['subscribed']
```

```python
x_train,x_test,y_train,y_test = train_test_split(mod_pc,y,test_size=0.2,random_state=42)
```

```python
lr = LogisticRegression()
```

```python
mod_pc
```

```python
mod_lr = lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
mod_lr.score(x_test,y_test)
```

```python
mat = m.confusion_matrix(y_test,y_pred)
mat
```

```python
print(m.classification_report(y_test,y_pred))
```

```python
m.roc_auc_score(y_test,mod_lr.predict_proba(x_test)[:,1])
```

### with WHiten = False

```python
pca = PCA(n_components=0.95)
```

```python
x_dum
```

```python
scaler = StandardScaler()
```
