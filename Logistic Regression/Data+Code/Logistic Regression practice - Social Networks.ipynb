```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
df
```

## Extracting Variables

```python
X = df.iloc[:,[2,3]].values
X
```

```python
Y = df.iloc[:,4].values
Y
```

## Heatmap to see Correlation

```python
sns.heatmap(df.iloc[:,[0,2,3,4]].corr(), annot=True, cmap = 'RdYlGn')
```

### Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3)
```

### Feature Scaling

```python
standard_scaler = StandardScaler()
x_train = standard_scaler.fit_transform(x_train)
x_test = standard_scaler.transform(x_test)
```

```python
x_test
```

### Instantiating and fitting the model to Training Dataset

```python
log_reg = LogisticRegression()
log_reg.fit(x_train,y_train)
```

### Prediction for Test Datset

```python
y_pred = log_reg.predict(x_test)
y_pred
```

```python
y_test
```

## Visualizing the training set Result

```python
x_set,y_set = x_train,y_train
x1,x2 = np.meshgrid(np.arange(start = x_set[:,0].min()-1, stop = x_set[:,0].max()+1, step = 0.01),
                   np.arange(start = x_set[:,1].min()-1, stop = x_set[:,1].max()+1, step = 0.01))
plt.contourf(x1,x2,log_reg.predict(np.array([x1.ravel(),x2.ravel()]).T).reshape(x1.shape),alpha=0.5,)

plt.xlim(x1.min(),x1.max())
plt.ylim(x2.min(),x2.max())
for i,j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set==j,0], x_set[y_set==j,1],
                c = ListedColormap(['blue','green'])(i), label=j)
    
plt.title('Logistic Regression (Train set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualizing the Test set Result

```python
X2Set,Y2Set = x_test,y_test

Xx,Yy = np.meshgrid(np.arange(X2Set[:,0].min()-1, X2Set[:,0].max()+1, 0.01),
                    np.arange(X2Set[:,1].min()-1, X2Set[:,1].max()+1, 0.01))

plt.contourf(Xx,Yy,log_reg.predict(np.array([Xx.ravel(),Yy.ravel()]).T).reshape(Xx.shape),
             alpha=0.75,cmap = ListedColormap(('black','blue')))

plt.xlim(Xx.min(),Xx.max())
plt.ylim(Yy.min(),Yy.max())
for i,j in enumerate(np.unique(Y2Set)):
    plt.scatter(X2Set[Y2Set==j,0], X2Set[Y2Set==j,1],
                c=ListedColormap(['yellow','red'])(i), label=j)
    
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
