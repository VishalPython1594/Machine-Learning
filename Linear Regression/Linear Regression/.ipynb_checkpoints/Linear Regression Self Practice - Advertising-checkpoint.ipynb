## The very first thing we need to check is that whether x and y are linearly related or not. Both in the linear and logistic regression. This is also or very first assumption behind these alogorithms.

```python
sns.pairplot(advertising,x_vars=['TV','Radio','Newspaper'], y_vars='Sales',size=5,kind='scatter')
```

```python
sns.heatmap(advertising.corr(),annot=True)
```

```python
x = advertising['TV']
y = advertising['Sales']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7, test_size=0.3)
```

### Adding a constant (to get an intercept)

```python
x_train_sm = sm.add_constant(x_train)
```

### Fitting the regression model using OLS (of SM)

```python
lr = sm.OLS(y_train,x_train_sm).fit()
```

```python
lr.params
```

```python
lr.summary()
```

### Here our Null Hypothesis would be - Np change will be there in the output using constant and slope of TV
### Alternative - Changes can be observed in the output with the change in the constant and slope of TV

```python
# Visualize the regression line - Best fit line
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train, 0.044617*x_train + 7.262150, 'r')
plt.show()
```

```python
# Error Analysis
```

```python
y_train_pred = lr.predict(x_train_sm)
res = (y_train-y_train_pred)
```

```python
res
```

```python
sns.distplot(res,bins=15)
plt.title('Error Terms')
plt.xlabel('Residual Errors')
plt.ylabel('freq')
```

```python
# Residual Plot
```

```python
plt.scatter(y_train_pred,res)
```

```python
# Adding Constant
```

```python
x_test_sm = sm.add_constant(x_test)
```

```python
y_test_pred = lr.predict(x_test_sm)
```

```python
res_test = (y_test - y_test_pred)
```

```python
# Plotting the histogram for errors
```
