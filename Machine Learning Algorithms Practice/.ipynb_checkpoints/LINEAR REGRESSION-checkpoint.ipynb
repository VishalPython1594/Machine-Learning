```python
x = df['TV']
y = df['Sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
import statsmodels.api as sm
```

```python
x_train_sm = sm.add_constant(x_train)
```

```python
lr = sm.OLS(y_train,x_train_sm).fit()
```

```python
lr.params
```

```python
lr.summary()
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train, 6.779253 + 0.049088*x_train, 'r')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.title('TV vs SALES')
plt.show()
```

```python
y_train_pred = lr.predict(x_train_sm)
error = y_train - y_train_pred
```

```python
sns.distplot(error,bins=15)
plt.title('Error Terms')
plt.xlabel('Residual Errors')
plt.ylabel('freq')
```

```python
plt.scatter(y_train_pred,error)
```

```python
#for Testing data
```

```python
x_test_sm = sm.add_constant(x_test)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, 6.779253 + 0.049088*x_test, 'r')
```

```python
y_test_pred = lr.predict(x_test_sm)
```

```python
error_test = y_test - y_test_pred
```

```python
sns.distplot(error_test,bins=15)
```

```python
plt.scatter(y_test_pred,error_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
rs = r2_score(y_test,y_test_pred)
rs
```

```python
#ADjusted R2
```

```python
adjr2 = 1-((1-rs)*(199)/(198))
adjr2
```

```python
mse = mean_squared_error(y_test,y_test_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
sales_residuals = lr.resid
sales_residuals
```

```python
probplot = sm.ProbPlot(sales_residuals)
probplot.ppplot(line='45')
plt.show()
```
