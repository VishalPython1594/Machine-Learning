```python
#Imports

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import statsmodels.api as sm

#Supress the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#Importing the Dataset

advertising = pd.read_csv("advertising.csv")
```

```python
advertising
```

```python
#Peeking the Data

advertising.shape
```

```python
advertising.info()
```

```python
advertising.describe()
```

```python
#Visualisation

sns.pairplot(advertising, x_vars = ['TV', 'Radio', 'Newspaper'], y_vars = 'Sales', size = 5, kind = 'scatter')
```

```python
sns.heatmap(advertising.corr(), cmap = "YlGnBu", annot = True)
plt.show()
```

```python
X = advertising['TV']
y = advertising['Sales']
```

```python
#Splitting the data into Train and Test (with split ratio = 70-30)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 5)
```

```python
X_train
```

```python
X_test.head()
```

```python
import statsmodels.api as sm
```

```python
#Adding a constant (to get an intercept)
X_train_sm = sm.add_constant(X_train)
```

```python
#Fitting the regression model using OLS (of SM)

lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
lr.params
```

```python
lr.summary()
```

```python
#Visualize the regression line - best fit line
plt.scatter(X_train, y_train)
plt.plot(X_train, 6.74 + 0.050 * X_train, 'r')
plt.show()
```

```python
#Error Analysis

y_train_pred = lr.predict(X_train_sm)

res = (y_train - y_train_pred)
```

```python
#PLotting the histogram of errors

sns.distplot(res, bins = 15)
plt.title('Error Terms')
plt.xlabel('Residual Errors')
plt.ylabel('freq')
```

```python
plt.scatter(y_train_pred, res)
```

```python
#Adding constant
X_test_sm = sm.add_constant(X_test)
```

```python
#Predicting
y_test_pred = lr.predict(X_test_sm)
```

```python
res_test = (y_test - y_test_pred)
```

```python
#PLotting the histogram of errors

sns.distplot(res_test, bins = 15)
plt.title('Test Data Error Terms')
plt.xlabel('Residual Errors')
plt.ylabel('freq')
```

```python
plt.scatter(y_test_pred,res_test)
```

```python
def standardize(vals):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(standardize(y_test_pred), standardize(res_test))
plt.title('Residuals vs Predicted Values')
plt.xlabel('Standardized Predicted Values')
plt.ylabel('Standardized Residuals')
```

```python
from sklearn.metrics import r2_score

r_sq = r2_score(y_test, y_test_pred)
r_sq
```

```python
#Splitting the data into Train and Test (with split ratio = 70-30)

from sklearn.model_selection import train_test_split
X_train_lm, X_test_lm, y_train_lm, y_test_lm = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 5)
```

```python
X_train_lm =  X_train_lm.values.reshape(-1,1)
X_test_lm = X_test_lm.values.reshape(-1,1)
```

```python
#Method 2
from sklearn.linear_model import LinearRegression

lm = LinearRegression()

lm.fit(X_train_lm, y_train_lm)
```

```python
print(lm.intercept_)
```

```python
print(lm.coef_)
```

```python
y_train_pred = lm.predict(X_train_lm)
y_test_pred = lm.predict(X_test_lm)
```

```python
print(r2_score(y_train_lm, y_train_pred))
```

```python
print(r2_score(y_test_lm, y_test_pred))
```

```python
#Assignment: Adj R2, TV, TV+Radio, Radio+Newspaper, TV+Newspaper (R2, Adj R2, )
# create X and y
feature_cols = ['TV', 'Radio', 'Newspaper']
X = advertising[feature_cols]
y = advertising.Sales

# instantiate and fit
lm2 = LinearRegression()
lm2.fit(X, y)

# print the coefficients
list(zip(feature_cols, lm2.coef_))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = []
for i in range(X.shape[1]):
    vif.append(variance_inflation_factor(X,i))
```

```python
pd.DataFrame({'VIF':vif},index = X.columns)
```
