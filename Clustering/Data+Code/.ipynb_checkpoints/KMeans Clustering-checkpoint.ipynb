## K Means Cluster

```python
from sklearn.cluster import KMeans
```

```python
x = df.iloc[:,3:]
```

```python
# To find the best K (Cluster) basis Elbow method

wcss = []
# Within the cluster sum of squared errors
for i in range(1,15):
    kmeans = KMeans(n_clusters=i,init = 'k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,15),wcss)
```

```python
wcss
```

```python
# As per elbow method, best k value is 5

kmeans = KMeans(n_clusters=5,init='k-means++')
mod_k = kmeans.fit_predict(x)
```

```python
kmeans.inertia_
```

```python
mod_k
```

```python
kmeans_classic = KMeans(n_clusters=5,init='k-means++')
kmeans_classic.fit(x)
```

```python
pred = kmeans_classic.predict(x)
pred
```

```python
combdf = pd.concat([df,pd.DataFrame(mod_k)],axis=1)
combdf
```

```python
# visualize the cluster

plt.scatter(x.values[mod_k==0,0], x.values[mod_k==0,1], s = 100, c = 'red', label='Cluster_0')
plt.scatter(x.values[mod_k==1,0], x.values[mod_k==1,1], s = 100, c = 'blue', label='Cluster_1')
plt.scatter(x.values[mod_k==2,0], x.values[mod_k==2,1], s = 100, c = 'yellow', label='Cluster_2')
plt.scatter(x.values[mod_k==3,0], x.values[mod_k==3,1], s = 100, c = 'pink', label='Cluster_3')
plt.scatter(x.values[mod_k==4,0], x.values[mod_k==4,1], s = 100, c = 'green', label='Cluster_4')
```
