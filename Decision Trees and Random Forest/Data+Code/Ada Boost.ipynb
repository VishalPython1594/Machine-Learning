```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('German_Credit_Data.csv')
df.head()
```

```python
x = df.drop('status',axis=1)
y = df['status']
```

```python
x = pd.get_dummies(x,drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
## Initializing Decision Tree Classifier to use as base classifier
dt_base_model = DecisionTreeClassifier()
```

```python
# Initializing AdaBoost classifier with 200 classifers
ada_model = AdaBoostClassifier(dt_base_model,n_estimators=200, random_state=42)
```

```python
# Fitting AdaBoost model to training set
ada_model.fit(x_train,y_train)
```

### Performance Measures - AdaBoost

```python
y_pred = ada_model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
metrics.roc_auc_score(y_test,ada_model.predict_proba(x_test)[:,1])
```

```python
mat = metrics.confusion_matrix(y_test,y_pred)
mat
```

```python
print(metrics.classification_report(y_test,y_pred))
```
